# Magic cleaner I:

```python
# Don't run it twice !!
df_recipes = df_recipes.astype(str)
for col in df_recipes.columns:
    if (col != "recipe_url")&(col!="meal") :
        df_recipes[col] = df_recipes[col].apply(lambda x: re.findall(r'(?<=>).*?(?=<)', x))
```

```python
df_recipes
```

# Magic Cleaner II:

```python
df_recipes = df_recipes.astype(str).replace("[]",np.NaN)
df_recipes = df_recipes.astype(str).replace("[^\w\s\.\,\/]"," ",regex=True)
```

```python
df_recipes = df_recipes[df_recipes.recipe_name != "nan"]
```

```python
df_recipes
```

## yield

```python
df_recipes["yield"][1]
```

```python
df_recipes["yield"] = df_recipes["yield"].astype(str)
```

```python
df_recipes["yield"] = df_recipes["yield"].astype(str).replace("(span)|(data)|(amount)","",regex=True)
df_recipes["yield"] = df_recipes["yield"].astype(str).replace("\s\s\s\s\d+","",regex=True)
df_recipes["yield"] = df_recipes["yield"].astype(str).replace("[^\d\,\w]","",regex=True)
df_recipes["yield"] = df_recipes["yield"].astype(str).replace("\,"," ",regex=True)
df_recipes["yield"] = df_recipes["yield"].astype(str).replace("\d\sx","",regex=True)
df_recipes["yield"] = df_recipes["yield"].astype(str).replace("\s{3,}","",regex=True)
```

```python
df_recipes["ingredient"] = df_recipes["ingredient"].astype(str).replace("\s\s\s\d+","",regex=True)
```

```python
df_recipes["ingredient"] = df_recipes["ingredient"].astype(str).replace("(xa0)","",regex=True)
df_recipes["ingredient"] = df_recipes["ingredient"].astype(str).replace("\s{2,}\,","",regex=True)
df_recipes["ingredient"] = df_recipes["ingredient"].astype(str).replace("\s{2,}"," ",regex=True)
df_recipes["ingredient"] = df_recipes["ingredient"].astype(str).replace("\s\/(li)","",regex=True)
df_recipes["ingredient"] = df_recipes["ingredient"].astype(str).replace("\s\/","",regex=True)
df_recipes["ingredient"] = df_recipes["ingredient"].astype(str).replace("\s\,","",regex=True)
```

```python
def random_menu_maker(days):
    random_df = pd.DataFrame(columns=["breakfast","apetizer","lunch","snack",
                                        "dinner","ingredient_list","total_cooking_time",
                                        "average_rating","kcal_sum","sugar_g_sum","sodium_g_sum","fat_g_sum",
                                        "saturated_fat_g_sum", "carbohydrates_g_sum","fiber_g_sum","protein_g_sum"])
    data = []
    #dictionary = {"breackfast":a,"apetizer":b,"lunch":c,"snack":d,
                 #"dinner":e,"ingredient_list":f,"total_cooking_time":g,
                 #"average_rating":h,"kcal_sum":i,"sugar_g_sum":j,"sodium_g_sum":k,"fat_g_sum":l,
                 #"saturated_fat_g_sum":m, "carbohydrates_g_sum":n,"fiber_g_sum":o,"protein_g_sum":p}
    
    for day in range(days):
        data = []
        A = df_recipes_clean[df_recipes_clean["meal"]=="breakfast"].sample(n=1)
        a = A[["recipe_name"]].values[0][0]
        
        B = df_recipes_clean[df_recipes_clean["meal"]=="snack"].sample(n=1)
        b = B[["recipe_name"]].values[0][0]
        
        C = df_recipes_clean[df_recipes_clean["meal"]=="main_dish"].sample(n=1)
        c = C[["recipe_name"]].values[0][0]
        
        D = df_recipes_clean[df_recipes_clean["meal"]=="breakfast"].sample(n=1) #im using breakfast as a snack for afternoon
        d = D[["recipe_name"]].values[0][0]
        
        E = df_recipes_clean[df_recipes_clean["meal"]=="main_dish"].sample(n=1)
        e = E[["recipe_name"]].values[0][0]
        
        #f = A["ingredient"]+B["ingredient"]+C["ingredient"]+D["ingredient"]+E["ingredient"]
        f = str(A[["ingredient"]].values[0][0]) +str(B[["ingredient"]].values[0][0]) + str(C[["ingredient"]].values[0][0])+ str(D[["ingredient"]].values[0][0])+str(E[["ingredient"]].values[0][0])
        
        g = float(sum(A["total_time_min"],
        sum(B["total_time_min"],
        sum(C["total_time_min"],
        sum(D["total_time_min"],
        E["total_time_min"])))))
        
        h = float((sum(A["rating"],
        sum(B["rating"],
        sum(C["rating"],
        sum(D["rating"],
        E["rating"])))))/5)
        
        i = int(sum(A["kcal"],
        sum(B["kcal"],
        sum(C["kcal"],
        sum(D["kcal"],
        E["kcal"])))))
        
        j = int(sum(A["sugar_g"],
        sum(B["sugar_g"],
        sum(C["sugar_g"],
        sum(D["sugar_g"],
        E["sugar_g"])))))

        k = int(sum(A["sodium_mg"],
        sum(B["sodium_mg"],
        sum(C["sodium_mg"],
        sum(D["sodium_mg"],
        E["sodium_mg"])))))
        
        l = float(sum(A["fat_g"],
        sum(B["fat_g"],
        sum(C["fat_g"],
        sum(D["fat_g"],
        E["fat_g"])))))
        
        m = float(sum(A["saturated_fat_g"],
        sum(B["saturated_fat_g"],
        sum(C["saturated_fat_g"],
        sum(D["saturated_fat_g"],
        E["saturated_fat_g"])))))
        
        n = float(sum(A["carbohydrates_g"],
        sum(B["carbohydrates_g"],
        sum(C["carbohydrates_g"],
        sum(D["carbohydrates_g"],
        E["carbohydrates_g"])))))
        
        o = float(sum(A["fiber_g"],
        sum(B["fiber_g"],
        sum(C["fiber_g"],
        sum(D["fiber_g"],
        E["fiber_g"])))))
        
        p = float(sum(A["protein_g"],
        sum(B["protein_g"],
        sum(C["protein_g"],
        sum(D["protein_g"],
        E["protein_g"])))))
        
        dictionary = {"breakfast":a,"apetizer":b,"lunch":c,"snack":d,
                 "dinner":e,"ingredient_list":f,"total_cooking_time":g,
                 "average_rating":h,"kcal_sum":i,"sugar_g_sum":j,"sodium_g_sum":k,"fat_g_sum":l,
                 "saturated_fat_g_sum":m, "carbohydrates_g_sum":n,"fiber_g_sum":o,"protein_g_sum":p}
        
        data.append(dictionary)
        random_df = pd.DataFrame(random_df.append(data,True))
    
    return random_df
        
```

# Building the recommender

```python
def vegan_recommender():
    print("Welcome to my daily menu recomnder. \n" 
          "It is 100% based on 'Danza de Fogones' recipes. \n"
          "It is 100% vegan! \n ")

    print("For how many days do you want to eat delicious food?")
    print("If you want a menu for just one day, can choose an ingredient that you want to use")

    days = input("Number of days: ")
    print("What are your daily nutritional requeriments?")

    if days == "0":
        print("You should try it at least one day!")
    #if days == "1":
        #print("\n\nPlease enter the ingredient that you want to use.")
        #ingredient_desired = input().lower()

    print("\n\nPlease enter the minimum amount of protein you would like to eat per day.")
    protein_desired = float(input("Minimun amount of protein in grams: "))
    if protein_desired > 60:
            protein_desired = 60
            print("\n\nThis was a high number!")
            print("We will suggest you some high-protein menus")
            print("You can have a double portion of some meals \n if you want to increase your protein intake ")

    print("\n\nPlease enter the minimum amount of kcal you would like to eat per day.")
    kcal_desired = float(input("Minimun amount of kcal: "))
    if protein_desired > 1500:
            protein_desired = 1500
            print("\n\n We will suggest you some high-kcal menus")
            print("You can have a double portion of some meals \n if you want to increase your kcal intake ")

    double_portion_days = []
    #for day in range(int(days)):
    RM = random_menu_maker(1)
    RM_2 = random_menu_maker(1)
    max_iter = 0
    while (len(RM) < int(days)+1) & (max_iter <5000):
        if  float(RM_2.protein_g_sum) >= float(protein_desired):
            if float(RM_2.kcal_sum) >= float(kcal_desired) :
                RM = pd.concat([RM,RM_2],axis=0)
                RM_2 = random_menu_maker(1)
        else:
            RM_2 = random_menu_maker(1)
        max_iter = max_iter + 1

    RM.reset_index()
    RM = RM.iloc[1:]
    RM.reset_index()
    return RM.reset_index()
```
