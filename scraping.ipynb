## Getting all the recipes from all the pages

```python
def scraping_snacks(pages_to_scrape):
    url_snacks_list = []
    for item in pages_to_scrape:
        url = item
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        
        for i in range(len(soup.select("article a"))):
            if i%2==0:
                url_snacks_list.append(soup.select("article a")[i]["href"])
    return url_snacks_list
    
        
```

```python
def scraper(url_list):

    recipe_name_list = []
    recipe_url_list = []
    date_list = []
    author_list = []
    cuisine_list = []
    description_list = []
    rating_list = []
    total_time_list = []
    yield_list = []
    #portion_size_list = []
    serving_size_list = []
    ingredient_list = [] 
    calories_list = []
    sugar_list = []
    sodium_list = []
    fat_list = []
    saturated_fat_list = []
    carbohydrates_list = []
    fiber_list = []
    protein_list = []
    
    for url in url_list:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        
   
        
        try:
            recipe_name_list.append(str(soup.select("div > header > h1")))
        except:
            recipe_name_list.append(str("NaN"))
        try:
            recipe_url_list.append(str(url))
        except:
            recipe_url_list.append(str("NaN"))
        try:
            author_list.append(str(soup.select("div.tasty-recipes-details ul span.tasty-recipes-author-name")))
        except:
            author_list.append(str("NaN"))
        try:
            cuisine_list.append(str(soup.select("div.tasty-recipes-details ul span.tasty-recipes-cuisine")))
        except:
            cuisine_list.append(str("NaN"))
        try:
            description_list.append(str(soup.select("div.tasty-recipes-description p")))
        except:
            description_list.append(str("NaN"))
        try:
            rating_list.append(str(soup.select("span.rating-label span.average"))) 
        except:
            rating_list.append(str("NaN"))
        try:
            total_time_list.append(str(soup.select("div.tasty-recipes-details ul span.tasty-recipes-total-time ")))
        except:
            total_time_list.append(str("NaN"))
        try:
            yield_list.append(str(soup.select("div.tasty-recipes-details ul span.tasty-recipes-yield")))
        except:
            yield_list.append(str("NaN"))
        #try:
            #portion_size_list.append()
        #except:
            #x=0
        try:
            serving_size_list.append(str(soup.select("span.tasty-recipes-serving-size")))
        except:
            serving_size_list.append(str("NaN"))
        try:
            ingredient_list.append(str(soup.select("div.tasty-recipes-ingredients-body ul li")))
        except:
            ingredient_list.append(str("NaN"))
        try:
            calories_list.append(str(soup.select("span.tasty-recipes-calories")))
        except:
            calories_list.append(str("NaN"))
        try:
            sugar_list.append(str(soup.select("span.tasty-recipes-sugar")))
        except:
            sugar_list.append(str("NaN"))
        try:
            sodium_list.append(str(soup.select("span.tasty-recipes-sodium")))
        except:
            sodium_list.append(str("NaN"))
        try:
            fat_list.append(str(soup.select("span.tasty-recipes-fat")))
        except:
            fat_list.append(str("NaN"))
        try:
            saturated_fat_list.append(str(soup.select("span.tasty-recipes-saturated-fat")))
        except:
            saturated_fat_list.append(str("NaN"))
        try:
            carbohydrates_list.append(str(soup.select("span.tasty-recipes-carbohydrates")))
        except:
            carbohydrates_list.append(str("NaN"))
        try:
            fiber_list.append(str(soup.select("span.tasty-recipes-fiber")))
        except:
            fiber_list.append(str("NaN"))
        try:
            protein_list.append(str(soup.select("span.tasty-recipes-protein")))       
        except:
            protein_list.append(str("NaN"))
        
    food_dict = {"recipe_name": recipe_name_list, 
                        "recipe_url":recipe_url_list ,
                        "author": author_list,
                        "cuisine":cuisine_list,
                        "description": description_list, 
                        "rating": rating_list,
                        "total_time": total_time_list,
                        "yield": yield_list ,
                        #"portion_size": portion_size_list,
                        "serving_size": serving_size_list,
                        "ingredient": ingredient_list,
                        "calories": calories_list,
                        "sugar": sugar_list,
                        "sodium": sodium_list,
                        "fat": fat_list,
                        "saturated_fat": saturated_fat_list,
                        "carbohydrates": carbohydrates_list,
                        "fiber": fiber_list ,
                        "protein": protein_list}
        
    return food_dict
```
